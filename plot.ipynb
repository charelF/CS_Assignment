```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("istherecorrelation.csv", sep=";")
```

```python
data.head()
```

```python
fig, ax = plt.subplots()

ax.set_xlabel("Year")
ax.plot(data["Year"], data["WO [x1000]"], color="blue", label="WO [x1000]")
ax.tick_params(colors="blue")

ax2 = ax.twinx()
ax2.plot(data["Year"], data["NL Beer consumption [x1000 hectoliter]"], color="orange", label="NL Beer consumption [x1000 hectoliter]")
ax2.tick_params(colors="orange")
fig.legend()
fig.show()

fig.savefig("pic", dpi=300)
```
